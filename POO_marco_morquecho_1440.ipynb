##### conceptos de POO (programacionorientada a objetos) en python

#### el prompt en IA gemini lenguaje Python, crear la clase
"animal" con los atributos "nombre", "raza" crear el contructor animal, y las funciones ladrar(), correr(), crear  los objetos pitbull, chihuahua, nivel principiante, no es interactivo, proporciona tu los datos

```python
# Definición de la clase
class Animal:
    # El constructor inicializa los atributos
    def __init__(self, nombre, raza):
        self.nombre = nombre
        self.raza = raza

    # Método para ladrar
    def ladrar(self):
        print(f"{self.nombre} dice: ¡Guau! ¡Guau!🐱‍👤")

    # Método para correr
    def correr(self):
        print(f"{self.nombre} está corriendo velozmente. 🐱‍🏍")

    def Morder(self):
        print(f"{self.nombre} está mordiendo a la vecina 🤦‍♀️")

# --- Creación de los objetos (instancias) ---

# Creamos el objeto Pitbull
pitbull = Animal("Rocco", "Pitbull")

# Creamos el objeto Chihuahua
chihuahua = Animal("Chispa", "Chihuahua")
# creamos el objeto perro callejero con datos
perrocallejero = Animal("solovino", "chafa")

# --- Uso de los objetos ---

print(f"Datos del primer perro: {pitbull.nombre} ({pitbull.raza})")
pitbull.ladrar()
pitbull.correr()

print("-" * 30)

print(f"Datos del segundo perro: {chihuahua.nombre} ({chihuahua.raza})")
chihuahua.ladrar()
chihuahua.correr()

print("-" * 30)

print(f"Datos del tercer perro: {perrocallejero.nombre} ({perrocallejero.raza})")
perrocallejero.ladrar()
perrocallejero.correr()
perrocallejero.Morder()
print(" programa creado por marco morquecho NL = 39")
```
